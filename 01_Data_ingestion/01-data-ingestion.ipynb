```python
print("\n RECURSIVE CHARCTEER TEXT SPLITTER")

import encodings
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import TextLoader
from torch._dynamo import source

dir_loader=DirectoryLoader(
    "data/text_files/",
    glob = "**/*.txt",## pattern 
    loader_cls=TextLoader,
    loader_kwargs={'encoding':'utf-8'},
    show_progress=True
)

documents = dir_loader.load()

print(f"📁 Loader {len(documents)} documents")
for i,doc in enumerate(documents):
    print(f"\n Document {i+1}: ")
    print(f" source {doc.metadata['source']}")
    print(f" Lenght: {len(doc.page_content)} character")
```

```python
# use recursive charcter text splitter to chunk from langchain_community.document_loaders.base_o365 import CHUNK_SIZE
from langchain_text_splitters import CharacterTextSplitter
from sympy.solvers.diophantine.diophantine import length
print("am using character text splitter")


recursive_splitter=CharacterTextSplitter(
    separator=' ',
    chunk_size = 200, # maximum ~200 charcheter
    chunk_overlap = 20, # Reapt the ~20 charcter in the next chunk
    length_function= len
)

split_doc = recursive_splitter.split_documents(documents)
```

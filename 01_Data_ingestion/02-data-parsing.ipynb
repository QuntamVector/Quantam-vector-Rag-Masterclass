```python
raw_pdf_text = """Company Financial Report


    The ﬁnancial performance for ﬁscal year 2024
    shows signiﬁcant growth in proﬁtability.
    
    
    
    Revenue increased by 25%.
    
The company's efﬁciency improved due to workﬂow
optimization.


Page 1 of 10
"""

def clean_text(text):
    """
    Clean the text by removing special characters,
    numbers, and converting to lowercase.
    """

    # Remove excessive whitespaces
    text = " ".join(text.split())

        # Fix ligatures
    text = text.replace("ﬁ", "fi")
    text = text.replace("ﬂ", "fl")

    return text
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
from langchain_core.documents import Document
from typing import List

class SmartPDFProcessor:
    """
    advanced pdf processing with error handling
    """

    def __init__(self,chunk_size=1000,chunk_overlap=100): ##(this is my constructor)
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter=RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap=chunk_overlap,
            separators=[" "]
        )
## these are constructor we are using for processing PDF in smart way

    def process_pdf(self,pdf_path:str)->List[Document]: # pdf path of string will return the List of Document
            # So here in order to use this Document we are importing from langchain_core.documents import Document
            """Process PDF with smart chunking and metadata enhancement"""
            # Lets load the pdf
            loader=PyPDFLoader(pdf_path)
            pages=loader.load()

            processed_chunks=[]

            for page_num,page in enumerate(pages):
                ## clean text
                cleaned_text=self._clean_text(page.page_content) ## Here _clean_text will define this function in the later stages

                # Skip nearly empty pages
                if len(cleaned_text.strip()) < 50:
                    continue

                # Create chunks with enhanced metadata
                chunks = self.text_splitter.create_documents(
                    texts=[cleaned_text],
                    metadatas=[{
                        **page.metadata,
                        "page": page_num + 1,
                        "total_pages": len(pages),
                        "chunk_method": "smart_pdf_processor",
                        "char_count": len(cleaned_text)
                    }]
                )
                
                processed_chunks.extend(chunks) # once you clean you are adding data to processed_chunks
                # here extend is method to add multiple lists
            return processed_chunks
    
        
#   
    def _clean_text(self, text: str) -> str:
        """Clean extracted text"""
        # Remove excessive whitespace
        text = " ".join(text.split())
        
        # Fix common PDF extraction issues
        text = text.replace("ﬁ", "fi")
        text = text.replace("ﬂ", "fl")
        
        return text
```

```python
type Vector = list[float]

def scale(scalar: float, vector: Vector) -> Vector:
    return [scalar * num for num in vector]

# passes type checking; a list of floats qualifies as a Vector.
new_vector = scale(2.0, [1.0, -4.2, 5.4])
```
